```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
taxas = pd.read_csv("taxa1.csv", sep = ',',decimal = '.')
```

```python
taxas.head()
```

```python
taxas.dtypes
```

```python
taxas.taxas = taxas.taxa.astype(float)
```

```python
taxas.periodo = pd.to_datetime(taxas.periodo, format='%d-%m-%y')
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.title('taxa de desemprego em % em 2020-21',fontsize = 16)
plt.plot(taxas.periodo,taxas.taxas,lw =3, label = 'desemprego', c = 'brown', linestyle = 'dashdot')
ax = plt.gca()
ax.set_facecolor('whitesmoke')
plt.ylabel("taxa em %",fontsize = 15)
plt.xlabel('data',fontsize = 15)
plt.xticks(rotation=30)
plt.ylim(11, 15)
plt.legend(loc = 'upper right')
```
