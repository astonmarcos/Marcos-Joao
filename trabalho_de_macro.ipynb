```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ipca = pd.read_csv(r'/content/ipca.csv', sep = ',',decimal = '.')
selic = pd.read_csv(r'/content/selic1.csv', sep = ',',decimal = '.')
real = pd.read_csv(r'/content/i_real.csv', sep = ',',decimal = '.')
nominal = pd.read_csv(r'/content/i_nominal.csv', sep = ',',decimal = '.')
```

```python
ipca.ipca = ipca.valor.astype(float)
nominal.nominal = nominal.valor.astype(float)
selic.selic = selic.valor.astype(float)
real.real = real.valor.astype(float)
```

```python
ipca.periodo = pd.to_datetime(ipca.periodo, format='%d-%m-%Y')
```

```python
ipca.dtypes
```

```python
ipca['taxa_real'] = selic.selic - ipca.ipca
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('i Juros de Pessoas Físicas, IPCA, Selic e i Real',fontsize = 16)
plt.plot(ipca.periodo,ipca.ipca,lw =3, label = ' IPCA', c = 'orange')
plt.plot(ipca.periodo,real.real,lw =3, label = 'i Juros Reais', c = 'purple')
plt.plot(ipca.periodo,selic.selic, lw = 3,label = 'Selic',c = 'Lime')
plt.plot(ipca.periodo,nominal.nominal, lw =3, label = 'Juros Pessoa Física', c = 'black')
plt.ylabel("Valor",fontsize = 15)
plt.xlabel('Anos',fontsize = 15)
plt.legend(loc = 'lower left')
```
