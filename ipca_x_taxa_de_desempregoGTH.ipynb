```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
ipca = pd.read_csv('ipca.csv', sep = ',',decimal = '.')
desemprego = pd.read_csv('taxadesemprego.csv', sep = ',',decimal = '.')
```

```python
ipca.head()
```

```python
desemprego.head()
```

```python
ipca.dtypes
```

```python
ipca.ipca = ipca.IPCA.astype(float)
```

```python
desemprego.dtypes
```

```python
ipca.periodo = pd.to_datetime(ipca.periodo, format='%d-%m-%Y')
```

```python
desemprego.desemprego = desemprego.taxa_De_desemprego.astype(float)
```

```python
ipca.dtypes
```

```python
desemprego.dtypes
```

```python
plt.rcParams['figure.figsize'] = (10,6)
plt.title('IPCA x taxa de desemprego em % em 2020-21',fontsize = 16)
plt.plot(ipca.periodo,ipca.IPCA,lw =3, label = 'ipca', c = 'purple', linestyle = 'solid')
plt.plot(ipca.periodo,desemprego.desemprego,lw =3, label = 'desemprego', c = 'orange', linestyle = 'solid')
ax = plt.gca()
ax.set_facecolor('lightgrey')
plt.xlabel('Per√≠odo',fontsize = 15)
plt.ylabel("Taxas em %",fontsize = 15)
plt.xticks(rotation=30)
plt.legend(loc = 'upper right')
```
