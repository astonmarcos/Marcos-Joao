```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cpi_countries= pd.read_csv(r'CPIYoY.csv' , sep = ',',decimal = '.')
```

```python
cpi_countries.head
```

```python
cpi_countries.dtypes
```

```python
plt.figure(figsize=(16,8))
plt.title('CPI Countries YoY%', fontsize = 12)
plt.yscale('log')
plt.bar(cpi_countries.country, cpi_countries.cpi_inflation_YoY, label=cpi_countries.cpi_inflation_YoY, color="darkblue")
for index,data in enumerate(cpi_countries.cpi_inflation_YoY):
    plt.text(x=index, y=data+1, s=f"{data}", fontdict=dict(fontsize=9), fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```
