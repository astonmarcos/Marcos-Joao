```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crescimentoBRxARG = pd.read_csv(r'crescimento_csv.csv', sep = ',',decimal = '.')
consumoBRxARG = pd.read_csv(r'csv_consumo.csv'  , sep = ',',decimal = '.')
```

```python
crescimentoBRxARG.head()
```

```python
consumoBRxARG.head()
```

```python
crescimentoBRxARG.periodo = pd.to_datetime(crescimentoBRxARG.periodo, format='%Y')
crescimentoBRxARG.crescimentoBR = crescimentoBRxARG.crescimentoBR.astype(float)
crescimentoBRxARG.crescimentoARG = crescimentoBRxARG.crescimentoARG.astype(float)
```

```python
consumoBRxARG.periodo = pd.to_datetime(consumoBRxARG.periodo, format='%Y')
consumoBRxARG.consumoBR = consumoBRxARG.consumo_familiaBR.astype(float)
consumoBRxARG.consumoARG = consumoBRxARG.consumo_familiaARG.astype(float)
```

```python
crescimentoBRxARG.dtypes
```

```python
consumoBRxARG.dtypes
```

LETRA A:

```python
plt.rcParams['figure.figsize'] = (12,7)
plt.title('Crescimento anual: Brasil x Argentina (Gross Capital Formation)',fontsize = 16)
plt.plot(crescimentoBRxARG.periodo,crescimentoBRxARG.crescimentoBR, lw=3 , label='Crescimento BR',c= 'green')
plt.plot(crescimentoBRxARG.periodo,crescimentoBRxARG.crescimentoARG, lw=3 , label='Crescimento ARG',c= 'lightskyblue')
plt.ylabel("Valor",fontsize = 15)
plt.xlabel('Anos',fontsize = 15)
plt.ylim(-20,38)
plt.legend(loc = 'upper right')
```

LETRA B

```python
plt.rcParams['figure.figsize'] = (12,7)
plt.title('Consumo: Brasil x Argentina (Household Final Consumption)',fontsize = 16)
plt.plot(consumoBRxARG.periodo,consumoBRxARG.consumo_familiaBR, lw=3 , label='Consumo familia BR',c= 'green')
plt.plot(consumoBRxARG.periodo,consumoBRxARG.consumo_familiaARG, lw=3, label= 'Consumo familia ARG', c='lightskyblue')
plt.ylabel("Valor",fontsize = 15)
plt.xlabel('Anos',fontsize = 15)
plt.ylim(-6,13 )
plt.legend(loc = 'upper right')
```
